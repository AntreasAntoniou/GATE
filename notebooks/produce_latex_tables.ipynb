```python
from typing import Dict, List
from pylatex import Document, Tabular


ensemble_series = 1
metric_mapping = {
    f"testing/ensemble_{ensemble_series}/accuracy_top_1-epoch-mean": "acc@1",
    f"testing/ensemble_{ensemble_series}/accuracy_top_1-epoch-mean": "acc@1",
    f"testing/ensemble_{ensemble_series}/accuracy_top_5-epoch-mean": "acc@5",
    f"testing/ensemble_{ensemble_series}/accuracy_top_5-epoch-mean": "acc@5",
    f"testing/ensemble_{ensemble_series}/auc-macro": "auc",
    f"testing/ensemble_{ensemble_series}/dice_loss-epoch-mean": "dice-loss",
    f"testing/ensemble_{ensemble_series}-/dice_loss-epoch-mean": "dice-loss",
    f"testing/ensemble_{ensemble_series}/image_to_text_accuracy-epoch-mean": "image-to-text-acc@1",
    f"testing/ensemble_{ensemble_series}/image_to_text_accuracy_top_5-epoch-mean": "image-to-text-acc@5",
    f"testing/ensemble_{ensemble_series}/loss-epoch-mean": "loss",
    f"testing/ensemble_{ensemble_series}/mIoU": "mIoU",
    f"testing/ensemble_{ensemble_series}/mae_loss-epoch-mean": "mae",
    f"testing/ensemble_{ensemble_series}/text_to_image_accuracy-epoch-mean": "text-to-image-acc@1",
    f"testing/ensemble_{ensemble_series}/text_to_image_accuracy_top_5-epoch-mean": "text-to-image-acc@5",
}

def generate_latex_table_with_mappings(model_list: List, data_dict: Dict, metric_mapping: Dict) -> str:
    dataset_headers = []
    metric_headers = []
    table_rows = []
    first = True
    model_list = [m for m in model_list if m is not None] 
    data_dict = {k: v for k, v in data_dict.items() if k is not None}
    # print(model_list, data_dict, metric_mapping)
    for model_name in sorted(model_list):
        
        if model_name not in data_dict:
            raise ValueError(f"Model {model_name} not found in data_dict")
        
        sorted_dataset_dict = sorted(data_dict[model_name].items(), key=lambda x: x[0])
        row = []
        
        for dataset_name, dataset_metric_dict in sorted_dataset_dict:
            for metric_name, metric_value in dataset_metric_dict.items():
                if metric_name in metric_mapping:
                    mapped_metric_name = metric_mapping[metric_name]
                    if first:
                        dataset_headers.append(f"{dataset_name}")
                        metric_headers.append(f"{mapped_metric_name}")
                    
                    row.append(f"{metric_value['value']:.3f}" if metric_value['value'] is not None else "-")
        
        table_rows.append(row)
        first = False
    
    doc = Document()
    table = Tabular("l" * (len(dataset_headers) + 1))
    table.add_hline()
    table.add_row(["Model"] + dataset_headers)
    table.add_hline()
    table.add_row(["Metric"] + metric_headers)
    table.add_hline()
    for model_name, row in zip(sorted(model_list), table_rows):
        table.add_row([model_name] + row)
        table.add_hline()
    doc.append(table)
    return doc.dumps_content()
                    
                    
                    
# Example usage
models_to_include = list(
    model_to_dataset_to_metrics.keys()
)  # Assuming 'updated_data' contains the JSON data
latex_table_with_datasets = generate_latex_table_with_mappings(
    models_to_include, standardised_model_to_dataset_to_metrics, metric_mapping
)
with open("latex_table_with_datasets.tex", "w") as f:
    f.write(latex_table_with_datasets)
```

```python
def convert_tex_to_csv(tex_file_path, csv_file_path):
    with open(tex_file_path, 'r') as tex_file:
        latex_table = tex_file.read()
    
    # Remove unwanted LaTeX commands
    latex_table = latex_table.replace("\\begin{tabular}", "").replace("\\end{tabular}", "").replace("\\", "").strip()
    latex_table = latex_table.replace("normalsize%", "").replace("{llllllllllllllllllllllllllllllllllllllllllllllllllllllllllll}", "").replace("hline%", "").strip()
    # latex_table = latex_table.replace("textbf{", "").replace("}", "").strip()
    latex_table = latex_table.replace("%", "").strip()

    # Replace & with , and \\ with \n
    latex_table = latex_table.replace("&", ",").replace("\\\\", "\n")
    
    latex_table = latex_table.replace("\n\n", "\n").strip()
    latex_table = latex_table.replace("{-}", "-").strip()

    # Write to a CSV file
    with open(csv_file_path, 'w') as csv_file:
        csv_file.write(latex_table)

# Usage:
convert_tex_to_csv('/disk/scratch_fast1/aantoni2/GATE/notebooks/latex_table_with_datasets.tex', '31012024-processed.csv')
```
