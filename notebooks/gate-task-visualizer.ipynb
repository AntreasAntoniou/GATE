```python
import plotly.graph_objects as go
sns.set_palette('colorblind')

# Create lists to store source, target, and value data
source = []
target = []
value = []

# Create a mapping of nodes to integers
node_map = {}
i = 0

for modality, dataset, task in zip(data["Modality"], data["Dataset"], data["Task"]):
    if modality not in node_map:
        node_map[modality] = i
        i += 1
    if dataset not in node_map:
        node_map[dataset] = i
        i += 1
    if task not in node_map:
        node_map[task] = i
        i += 1

    source.append(node_map[dataset])
    target.append(node_map[task])
    value.append(1)
    
    source.append(node_map[modality])
    target.append(node_map[dataset])
    value.append(1)



fig = go.Figure(data=[go.Sankey(
    node = dict(
    #   pad = 15,
      thickness = 20,
      line = dict(color = "black", width = 0.5),
      label = list(node_map.keys()),
      
    ),
    link = dict(
      source = target,
      target = source,
      value = value
  ))])

fig.update_layout(
    title_text="GATE Task Overview", 
    font_size=16,  # Increase font size
    autosize=True,  # Disable automatic sizing
    width=800,  # Set figure width
    height=1200,  # Set figure height
)

fig.show()
```
